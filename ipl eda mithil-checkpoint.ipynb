```python
# 28. The winners in each Season

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'matches.csv' is the CSV file containing the IPL match data
df = pd.read_csv('matches.csv')

# Get the count of matches won by each team in each season
season_winner_counts = df.groupby(['season', 'winner'])['id'].count().reset_index()

# Pivot the data to create a matrix for stacked bar chart
winner_matrix = season_winner_counts.pivot('season', 'winner', 'id').fillna(0)

# Create a stacked bar chart
plt.figure(figsize=(12, 6))
winner_matrix.plot(kind='bar', stacked=True, colormap='Paired')
plt.title('Number of Matches Won by Each Team in Each Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches Won')
plt.legend(title='Winner', bbox_to_anchor=(1, 1))
plt.xticks(rotation=45)
plt.show()


```

```python
# 31. The toss winner, toss decision, winner in final matches.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'matches.csv' is the CSV file containing the IPL match data
df = pd.read_csv('matches.csv')

# Get the count of toss winners, toss decisions, and match winners
toss_info = df.groupby(['toss_winner', 'toss_decision', 'winner'])['id'].count().reset_index()

# Create a grouped bar chart
plt.figure(figsize=(12, 6))
sns.barplot(x='toss_winner', y='id', hue='toss_decision', data=toss_info, ci=None)
sns.barplot(x='toss_winner', y='id', hue='winner', data=toss_info, ci=None, alpha=0.7)
plt.title('Toss Winner, Decision, and Match Winner')
plt.xlabel('Toss Winner')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.legend(title='Toss Decision', bbox_to_anchor=(1, 1))
plt.show()










```

```python
# 32. The man of the match for each winning team

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'matches.csv' is the CSV file containing the IPL match data
df = pd.read_csv('matches.csv')

# Get the count of matches won by each team in each season
man_of_match_info = df.groupby(['winner', 'player_of_match'])['id'].count().reset_index()

# Pivot the data to create a matrix for heatmap
man_of_match_matrix = man_of_match_info.pivot('winner', 'player_of_match', 'id').fillna(0)

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(man_of_match_matrix, cmap='YlGnBu', annot=True, fmt='g', linewidths=0.5)
plt.title('Number of Times Each Player Became Man of the Match for Each Winning Team',fontsize=19)
plt.xlabel('Player of the Match')
plt.ylabel('Winning Team')
plt.xticks(rotation=90)
plt.show()



```

```python
# 34. What is the probability of winning a match if the toss was won?

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'matches.csv' is the CSV file containing the IPL match data
df = pd.read_csv('matches.csv')

# Check if toss winner is the same as the match winner
df['toss_match_winner'] = df['toss_winner'] == df['winner']

# Create a countplot for visualization
plt.figure(figsize=(8, 6))
sns.countplot(x='toss_match_winner', data=df)
plt.title('Matches where Toss Winner is the Same as Match Winner')
plt.xlabel('Toss Winner Same as Match Winner')
plt.ylabel('Count')
plt.xticks(ticks=[0, 1], labels=['No', 'Yes'])
plt.show()
```

```python
# 35. The top 10 leading run scorers in IPL

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming 'matches.csv' is the CSV file containing the IPL match data
df1=pd.read_csv("C:/Users/ayush/Desktop/deliveries.csv")

# Get the top 10 leading run scorers in IPL
top_10_run_scorers = df1.groupby(['batsman'])['batsman_runs'].sum().sort_values(ascending=False).head(10)

# Create a bar plot
plt.figure(figsize=(8, 4))
sns.barplot(x=top_10_run_scorers.index, y=top_10_run_scorers.values, palette='viridis')
plt.title('Top 10 Leading Run Scorers in IPL')
plt.xlabel('Player')
plt.ylabel('Total Runs')
plt.xticks(rotation=90)
plt.show()
```
